sharpening eczema

```python
import os

import cv2

# Function to sharpen an image
def sharpen_image(img):
    # Define new dimensions for resizing


    # Apply Gaussian blur
    gaussian_blur = cv2.GaussianBlur(img, (7, 7), 2)

    # Perform image sharpening
    sharpened_img = cv2.addWeighted(img, 1.5, gaussian_blur, -0.5, 0)

    return sharpened_img

# Directory containing the images
input_directory = '/content/drive/MyDrive/Projects/SkinDiseaseDetectionSystem/Splited dataset/train/eczema'

# Output directory for sharpened images
# Ooutput_directory = '/content/drive/MyDrive/Projects/SkinDiseaseDetectionSystem/sharpened_data/train/eczema'

# List all image files in the input directory
image_files = [f for f in os.listdir(input_directory) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Process and sharpen each image and save to the same directory
for image_index, img_file in enumerate(image_files):
  # Load an image
  img_path = os.path.join(input_directory, img_file)
  img = cv2.imread(img_path)

  # Sharpen the image
  sharpened_img = sharpen_image(img)

  # Display the images
  #cv2.imshow('Original', img)
  #cv2.imshow('Sharpened', sharpened_img)

  # Wait for a key press
  #key = cv2.waitKey(0)

  # Save the sharpened image if 's' is pressed
  #if key == ord('s'):

  #define base name
  base_new_filename = 'eczema'

  #file renameing
  old_filepath = os.path.join(input_directory, img_file)
  file_extension = os.path.splitext(img_path)[1]
  new_filename = f'{base_new_filename}_{str(image_index + 1).zfill(4)}{file_extension}'
  new_filepath = os.path.join(input_directory, new_filename)

  #save to the same directory
  os.rename(old_filepath, new_filepath)
  print(f"Image renamed as: {new_filepath}")


# Close all OpenCV windows
cv2.destroyAllWindows()
```

**Rotate images**

**eczema**

```python
import cv2
import os

#directories
input_folder = '/content/drive/MyDrive/Projects/SkinDiseaseDetectionSystem/Splited dataset/train/eczema'
output_folder = '/content/drive/MyDrive/Projects/SkinDiseaseDetectionSystem/Dataset/train/eczema'

def process_and_save_images(input_folder, output_folder):

  # Create the output folder if it doesn't exist
  if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate through all image files in the input folder
for filename in os.listdir(input_folder):
  if filename.endswith(".png") or filename.endswith(".jpg"):
    # Read the image
    img_path = os.path.join(input_folder, filename)
    img = cv2.imread(img_path)

    # Process the image
    rotated_img_1 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    rotated_img_2 = cv2.rotate(rotated_img_1, cv2.ROTATE_90_CLOCKWISE)
    rotated_img_3 = cv2.rotate(rotated_img_2, cv2.ROTATE_90_CLOCKWISE)
    flipped_img_h = cv2.flip(img, 1)  # Horizontal flip
    flipped_img_v = cv2.flip(img, 0)  # Vertical flip

    # Save the processed images
    cv2.imwrite(os.path.join(output_folder, f"{filename}_rotated_90_1.jpg"), rotated_img_1)
    cv2.imwrite(os.path.join(output_folder, f"{filename}_rotated_90_2.jpg"), rotated_img_2)
    cv2.imwrite(os.path.join(output_folder, f"{filename}_rotated_90_3.jpg"), rotated_img_3)
    cv2.imwrite(os.path.join(output_folder, f"{filename}_flipped_h.jpg"), flipped_img_h)
    cv2.imwrite(os.path.join(output_folder, f"{filename}_flipped_v.jpg"), flipped_img_v)

    print(f"Processed and saved images for {filename}")

process_and_save_images(input_directory, output_directory)
```
